# <h1 style="font-family: Trebuchet MS; padding: 20px; font-size: 40px; color: #EF6262; text-align: center; line-height: 0.55;background-color: #262322"><b>Outlier treatment</b><br></h1>

```python
def remove_outliers(data):
    data_out = data.copy()
    for column in data_out.columns:
        if column != 'Class':
            Q1 = data_out[column].quantile(0.25)
            Q3 = data_out[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            data_out = data_out[(data_out[column] >= lower_bound) & (data_out[column] <= upper_bound)]
    
    return data_out
```

```python
df_out = df.copy()
df_out = remove_outliers(df)
df_out
```

# <h1 style="font-family: Trebuchet MS; padding: 20px; font-size: 40px; color: #EF6262; text-align: center; line-height: 0.55;background-color: #262322"><b>Data Balancing</b><br></h1>

- In order to handle imbalanced data, there are 2 options :

    - **Undersampling** : Decreases the majority samples of the target variable.
    - **Oversampling**  : Increase the minority samples of the target variable to the majority samples.
    
    
- For best performances, we will use oversampling because we don't want to use any important information & our minority class is too low.
- For data balancing, we will use **imblearn**.

```python
from sklearn.model_selection import train_test_split

x = df.drop('Class', axis=1)
y = df['Class']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

# Apply SMOTE to the training data
smote = SMOTE(sampling_strategy='auto', random_state=42)
under = RandomUnderSampler(random_state=42)

x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
x_train_resampled2, y_train_resampled2 = under.fit_resample(x_train, y_train)
```

# <h1 style="font-family: Trebuchet MS; padding: 20px; font-size: 40px; color: #EF6262; text-align: center; line-height: 0.55;background-color: #262322"><b>Modeling</b><br></h1>

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import RocCurveDisplay
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.metrics import precision_recall_curve
```

```python
def model(classifier, x_train, y_train, x_test, y_test):
    classifier.fit(x_train, y_train)
    prediction = classifier.predict(x_test)
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=42)
    print("Cross Validation Score:", '{0:.2%}'.format(cross_val_score(classifier, x_train, y_train, cv=cv, scoring='roc_auc').mean()))
    print("ROC_AUC Score:", '{0:.2%}'.format(roc_auc_score(y_test, prediction)))

    # Classification Report
    print(classification_report(y_test, prediction))
    
    # Create a new figure for ROC curve and confusion matrix
    fig, ax = plt.subplots(1, 2, figsize=(12, 5))

    # First subplot - ROC Curve
    roc_display = RocCurveDisplay.from_estimator(classifier, x_test, y_test, ax=ax[0])
    ax[0].set_title('ROC_AUC Plot')

    # Second subplot - Confusion Matrix
    confusion_matrix = metrics.confusion_matrix(y_test, prediction)
    accuracy = accuracy_score(y_test, prediction)
    misclassification_rate = 1 - accuracy

    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=['Fraud', 'Legit'])
    cm_display.plot(cmap=plt.cm.Blues, ax=ax[1])
    
    # Add accuracy and misclassification rate to the confusion matrix display using xlabel
    ax[1].set_xlabel(f'Predicted label \nAccuracy: {accuracy:.2%}  Misclass Rate: {misclassification_rate:.2%}')
    ax[1].set_title('Confusion Matrix')

    fig.tight_layout(w_pad=5)

    plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

classifier_lr = LogisticRegression()

model(classifier_lr, x_train_resampled, y_train_resampled, x_test, y_test)
```
